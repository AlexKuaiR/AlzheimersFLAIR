```python
import pydicom
import os
import numpy as np
import cv2
import pandas as pd
import numpy as np
import nibabel as nib
```

```python
import dicom2nifti

entries = os.listdir('ADNI-orig')

for s in entries:
    try:
        path = os.path.join('ADNI-orig', s)
        
        if ".DS_Store" not in os.path.basename(path):
            # subject id path is valid
            path = os.path.join(path, os.listdir(path)[0])
            # path.basename should be either Axial_FLAIR or AXIAL_FLAIR
        else:
            print(path, "subject id path not valid, contains DS_Store")
            continue
        
        all_paths = [] # because of branching from the dates
        if ".DS_Store" not in os.path.basename(path):
            # multiple dates
            for dates in os.listdir(path):
                date_path = os.path.join(path, dates)
                all_paths.append(date_path)
            
            #print(all_paths)
        else:
            print(path, "image description path not valid, contains DS_Store")
            continue
        
        for i in range(len(all_paths)):
            if ".DS_Store" not in os.path.basename(all_paths[i]):
                # description path is valid
                #print(all_paths[1])
                all_paths[i] = os.path.join(all_paths[i], os.listdir(all_paths[i])[0])
            else:
                print(all_paths[i], "date path not valid, contains DS_Store")
                continue
        
        count = 1
        for p in all_paths:
            if ".DS_Store" not in os.path.basename(p):
                # date path is valid
                print(p, "successful")
                filename = s + 't' + str(count) + '.nii.gz'
                count += 1
                dicom2nifti.dicom_series_to_nifti(p, os.path.join('ADNI-nifti', filename))
            else:
                print(p, "thing after date path not valid, contains DS_Store")
        
        #dicom2nifti.dicom_series_to_nifti(s, os.path.join(path_out, patient + '.nii.gz'))
    except dicom2nifti.exceptions.ConversionValidationError as e:
        print(f"Error occurred for patient: {patient}. Details: {e}")
        continue
    #break
    
```
