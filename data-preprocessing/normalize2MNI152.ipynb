```python
import os
import subprocess
import numpy as np
print(np.shape(os.listdir('filtered-bet')))
for s in os.listdir('filtered-bet'):
    if (s.endswith('nii.gz')):
        path = os.path.join('filtered-bet', s)
        template = 'fsl/data/standard/MNI152_T1_1mm_brain.nii.gz'
        out_path = os.path.join('standard-space2', s[:-7] + '.nii.gz')
        
        transformation_matrix = os.path.join('transform_mat', s + '_transformation_matrix')
        
        subprocess.run(['flirt', '-in', path, '-ref', template, '-out', out_path, '-omat', transformation_matrix,
                       '-cost', 'mutualinfo', '-dof', '6'])
        #command = 'flirt -in ' + path + ' -ref ' + template + ' -out ' + out_path + ' -omat transformation_matrix'
        #print(command)
        #subprocess.run(command, shell = True)
```
