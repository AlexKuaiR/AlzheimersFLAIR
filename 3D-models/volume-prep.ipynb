```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import skimage.io
import os 
import tqdm
import glob
import shutil

labels = pd.read_csv('AlzheimersFLAIR/patient_labels.csv')
labels
```

```python
## import numpy as np
import nibabel as nib
from skimage import exposure
import os

nifti_directory = 'AlzheimersFLAIR/data-preprocessing/n4-corrected2'
#print(len(nifti_directory))
patient_array = []
diagnosis = []

used_patients = []

for s in os.listdir(nifti_directory):
    if (s.endswith('.nii.gz')):
        
        cur_path = os.path.join(nifti_directory, s)
        
        curDiagnosisStr = labels.loc[labels['Subject ID'] == s[:10], 'Research Group'].values[0]
        
        research_group = os.path.join(curDiagnosisStr, s)
        
        out_path = os.path.join('brain-volumes', research_group)
        
        shutil.copy(cur_path, out_path)
        
        if os.path.exists(out_path):
            print(f"File '{s}' was successfully duplicated to '{out_path}'.")
        else:
            print("Failed to duplicate the file.")
```
